```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
file_path = '/Users/ahmad/Documents/NERC Internship/Excel/Actual Data/Dust/2:8:24/Data_HIRE4573_1722639792113 - Test Site.xlsx'
df = pd.read_excel(file_path, sheet_name='Data_HIRE4573_1722639792113', engine='openpyxl')

# Calculate max PM values
max_pm10 = df['PM10 particles (ug/m^3)'].max()
max_pm25 = df['PM2.5 particles (ug/m^3)'].max()
max_pm1 = df['PM1 particles (ug/m^3)'].max()

# Calculate average PM values
avg_pm10 = df['PM10 particles (ug/m^3)'].mean()
avg_pm25 = df['PM2.5 particles (ug/m^3)'].mean()
avg_pm1 = df['PM1 particles (ug/m^3)'].mean()

# Calculate L90, L50, and L10 for PM values
L90_pm10 = df['PM10 particles (ug/m^3)'].quantile(0.90)
L50_pm10 = df['PM10 particles (ug/m^3)'].quantile(0.50)
L10_pm10 = df['PM10 particles (ug/m^3)'].quantile(0.10)

L90_pm25 = df['PM2.5 particles (ug/m^3)'].quantile(0.90)
L50_pm25 = df['PM2.5 particles (ug/m^3)'].quantile(0.50)
L10_pm25 = df['PM2.5 particles (ug/m^3)'].quantile(0.10)

L90_pm1 = df['PM1 particles (ug/m^3)'].quantile(0.90)
L50_pm1 = df['PM1 particles (ug/m^3)'].quantile(0.50)
L10_pm1 = df['PM1 particles (ug/m^3)'].quantile(0.10)

plt.figure(figsize=(40, 20))

# Plot PM values
plt.plot(df['Date Time'], df['PM10 particles (ug/m^3)'], linewidth=3, label=f'PM10 particles (ug/m続) - Max: {max_pm10}', color="r", marker='o')
plt.plot(df['Date Time'], df['PM2.5 particles (ug/m^3)'], linewidth=3, label=f'PM2.5 particles (ug/m続) - Max: {max_pm25}', color="b", marker='o')
plt.plot(df['Date Time'], df['PM1 particles (ug/m^3)'], linewidth=3, label=f'PM1 particles (ug/m続) - Max: {max_pm1}', color="y", marker='o')

# Plot average lines
plt.axhline(y=avg_pm10, color='r', linestyle='--', linewidth=2, label=f'Average PM10: {avg_pm10:.2f}')
plt.axhline(y=avg_pm25, color='b', linestyle='--', linewidth=2, label=f'Average PM2.5: {avg_pm25:.2f}')
plt.axhline(y=avg_pm1, color='y', linestyle='--', linewidth=2, label=f'Average PM1: {avg_pm1:.2f}')

# Ln lines
plt.axhline(y=L90_pm10, color='r', linestyle='-.', linewidth=2, label=f'L90 PM10: {L90_pm10:.2f}')
plt.axhline(y=L50_pm10, color='r', linestyle=':', linewidth=2, label=f'L50 PM10: {L50_pm10:.2f}')
plt.axhline(y=L10_pm10, color='r', linestyle='-', linewidth=2, label=f'L10 PM10: {L10_pm10:.2f}')

plt.axhline(y=L90_pm25, color='b', linestyle='-.', linewidth=2, label=f'L90 PM2.5: {L90_pm25:.2f}')
plt.axhline(y=L50_pm25, color='b', linestyle=':', linewidth=2, label=f'L50 PM2.5: {L50_pm25:.2f}')
plt.axhline(y=L10_pm25, color='b', linestyle='-', linewidth=2, label=f'L10 PM2.5: {L10_pm25:.2f}')

plt.axhline(y=L90_pm1, color='y', linestyle='-.', linewidth=2, label=f'L90 PM1: {L90_pm1:.2f}')
plt.axhline(y=L50_pm1, color='y', linestyle=':', linewidth=2, label=f'L50 PM1: {L50_pm1:.2f}')
plt.axhline(y=L10_pm1, color='y', linestyle='-', linewidth=2, label=f'L10 PM1: {L10_pm1:.2f}')

#plt.suptitle('FIGURE 1', fontsize=40, y=1)
plt.title('Test Site Dust Data for 02/08/24', fontsize=30, pad=30)

plt.xlabel('Time', fontsize=30, labelpad = 30)
plt.ylabel('PM (ug/m続)', fontsize=30, labelpad = 30)
plt.legend(fontsize=25, loc='upper right', bbox_to_anchor=(1.2, 1))

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

plt.xlim(df['Date Time'].min(), df['Date Time'].min() + pd.Timedelta(days=1))
plt.ylim(0, df[['PM10 particles (ug/m^3)', 'PM2.5 particles (ug/m^3)', 'PM1 particles (ug/m^3)']].max().max() + 15)

hour_locator = mdates.HourLocator(interval=1)
plt.gca().xaxis.set_major_locator(hour_locator)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

plt.gca().yaxis.set_major_locator(plt.MultipleLocator(20))
plt.gca().yaxis.set_minor_locator(plt.MultipleLocator(2))

plt.tight_layout()
plt.show()
```
